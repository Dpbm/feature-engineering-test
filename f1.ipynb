```python
from sklearn.model_selection import train_test_split

features = ["battery_power","blue", "clock_speed", "dual_sim", "fc", "four_g", "int_memory", "m_dep", "mobile_wt", "n_cores", "pc", "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time", "three_g", "touch_screen", "wifi"]
target = "price_range"

x_train, x_test, y_train, y_test = train_test_split(train_df[features], train_df[target], random_state=45)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
def evaluate(predictions, answers):
    total_correct = 0
    total_wrong = 0
    total_predictions = len(predictions)
    for index, prediction in enumerate(predictions):
        if(answers[index] == prediction):
            total_correct += 1
            #print(f"{index} Correct")
        else:
            total_wrong += 1
            #print(f"{index} Wrong")
    print(f"Total Correct: {total_correct}")
    print(f"Total Wrong: {total_wrong}")
    print(f"Total Correct %: {(total_correct/total_predictions)*100}")
    print(f"Total Wrong %: {(total_wrong/total_predictions) * 100}")
```

```python
evaluate(predictions, y_test.tolist())
```

```python
from sklearn.feature_selection import mutual_info_classif
mi = mutual_info_classif(train_df[features], train_df[target], discrete_features=(train_df[features].dtypes == int)) 
mi_data = pd.Series(data=mi, index=train_df[features].columns, name="MI Scores")
mi_data.sort_values(ascending=False)
```

```python
features = ["battery_power","ram", "px_height", "px_width", "mobile_wt"]
x_train, x_test, y_train, y_test = train_test_split(train_df[features], train_df[target], random_state=45)
model.fit(x_train, y_train)
predictions = model.predict(x_test)
evaluate(predictions, y_test.tolist())
```

```python
output_predictions = model.predict(train_df[features])
test_df['price_range'] = pd.Series(output_predictions)
test_df.to_csv("final.csv", index=False)
```
